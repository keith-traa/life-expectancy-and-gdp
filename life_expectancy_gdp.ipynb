```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("all_data.csv")
print(df.head())
```

```python
print(pd.unique(df.Country))
```

```python
df_chile = df[df.Country == "Chile"]
df_china = df[df.Country == "China"]
df_germany = df[df.Country == "Germany"]
df_mexico = df[df.Country == "Mexico"]
df_usa = df[df.Country == "United States of America"]
df_zimbabwe = df[df.Country == "Zimbabwe"]
```

```python
fig, axs = plt.subplots(2, 3, figsize=(16,10))
plt.legend(pd.unique(df.Country))
axs[0][0] = plt.subplot(2,3,1)
axs[0][0].plot(df_chile["Life expectancy at birth (years)"], df_chile.GDP, 'r.')
axs[0][1] = plt.subplot(2,3,2)
axs[0][1].plot(df_china["Life expectancy at birth (years)"], df_china.GDP, 'b.')
axs[0][2] = plt.subplot(2,3,3)
axs[0][2].plot(df_germany["Life expectancy at birth (years)"], df_germany.GDP, 'g.')
axs[1][0] = plt.subplot(2,3,4)
axs[1][0].plot(df_mexico["Life expectancy at birth (years)"], df_mexico.GDP, 'c.')
axs[1][1] = plt.subplot(2,3,5)
axs[1][1].plot(df_usa["Life expectancy at birth (years)"], df_usa.GDP, 'k.')
axs[1][2] = plt.subplot(2,3,6)
axs[1][2].plot(df_zimbabwe["Life expectancy at birth (years)"], df_zimbabwe.GDP, 'm.')
plt.show()
```
